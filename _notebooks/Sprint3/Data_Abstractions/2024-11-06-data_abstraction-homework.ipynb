# JavaScript Car Class Code - Optional Homework (Extra Credit)

Create a `Car` class that models a car with a few basic properties and methods. Follow these steps:

### 1. **Initialize Properties**:
- Create properties for `make`, `model`, and `fuelLevel` (a number between 0 and 100 representing the amount of fuel in the car).
  
### 2. **Methods**:
- `drive(distance)`: Decreases the `fuelLevel` by a certain percentage based on the distance traveled.
- `refuel(amount)`: Increases the `fuelLevel` by a specified amount, but make sure that it does not exceed 100 (the max fuel capacity).
- `getFuelLevel()`: Returns the current `fuelLevel`.

### 3. **Encapsulation**:
- Keep the `fuelLevel` private, and only allow it to be modified or read via the `drive`, `refuel`, and `getFuelLevel` methods.
- Do not allow direct access to `fuelLevel` from outside the class.

### 4. **Validation**:
- Ensure that when refueling, the `fuelLevel` does not exceed 100.
- When driving, ensure that the car does not drive without enough fuel (so if fuel is 0, they cant drive).

### 5. **Testing**:
- Create a `Car` instance with an initial fuel level.
- Test the following:
  - Drive the car and check how the fuel level changes.
  - Refuel the car and check that the fuel level increases without exceeding 100.
  - Try driving the car when the fuel level is 0 and ensure that it cannot drive.

---

### Example Scenario:

1. **Create a Car instance**:
   - The car has a `make` of `"Toyota"`, a `model` of `"Corolla"`, and a `fuelLevel` of 50%.

2. **Test the following:**
   - Drive the car for 100 km (decreases fuel level by 10%).
   - Refuel the car with 30 units (increases fuel level, but should not exceed 100%).
   - Try to drive the car with 0 fuel and verify it doesnâ€™t drive.

---



```python
%%js 
class Car {
    constructor(make, model, fuelLevel = 100) {
        this.make = make;
        this.model = model;
        
        // Encapsulate fuel level
        let _fuelLevel = fuelLevel;

        // Get current fuel level
        this.getFuelLevel = function() {
            return _fuelLevel;
        };

        // Drive the car, decreasing fuel level based on distance
        this.drive = function(distance) {
            if (_fuelLevel <= 0) {
                return "Cannot drive; fuel level is at 0.";
            }

            // Assuming fuel consumption rate is 0.1% per kilometer
            const fuelConsumption = distance * 0.1;
            if (_fuelLevel - fuelConsumption >= 0) {
                _fuelLevel -= fuelConsumption;
                return `Drove ${distance} km. Fuel level is now ${_fuelLevel.toFixed(2)}%.`;
            } else {
                _fuelLevel = 0;
                return "Ran out of fuel during the drive.";
            }
        };

        // Refuel the car, ensuring fuel level does not exceed 100
        this.refuel = function(amount) {
            if (amount <= 0) {
                return "Refuel amount must be positive.";
            }
            _fuelLevel = Math.min(_fuelLevel + amount, 100);
            return `Refueled ${amount} units. Fuel level is now ${_fuelLevel.toFixed(2)}%.`;
        };
    }
}


const myCar = new Car("Toyota", "Corolla", 50);

console.log(myCar.drive(100)); 
console.log(myCar.refuel(30)); // Expected: Refueled 30 units. Fuel level is now 70.00%.

myCar.drive(500); // Consumes the remaining fuel
console.log(myCar.drive(10)); // Expected: Cannot drive; fuel level is at 0.

// Display the current fuel level
console.log("Final Fuel Level:", myCar.getFuelLevel()); // Expected: 0
```

```python
%%js
class BankAccount {
    constructor(initialBalance = 100) {
        let balance = initialBalance;


        this.deposit = function(amount) {
            if (amount > 0) {
                balance += amount;
                return `Deposited $${amount}. New balance: $${balance}`;
            } else {
                return "Deposit amount must be positive.";
            }
        };

        this.withdraw = function(amount) {
            if (amount > 0 && amount <= balance) {
                balance -= amount;
                return `Withdrew $${amount}. New balance: $${balance}`;
            } else if (amount > balance) {
                return "Insufficient funds for this withdrawal.";
            } else {
                return "Withdrawal amount must be positive.";
            }
        };

        this.getBalance = function() {
            return balance;
        };

        this.transfer = function(amount, targetAccount) {
            if (amount > 0 && amount <= balance) {
                this.withdraw(amount);
                targetAccount.deposit(amount);
                return `Transferred $${amount} to target account. New balance: $${balance}`;
            } else {
                return "Insufficient funds for this transfer.";
            }
        };
    }
}

class PremiumBankAccount extends BankAccount {
    constructor(initialBalance) {
        super(initialBalance);
    }

    
    withdraw(amount) {
        if (amount > 0) {
            return super.withdraw(amount);
        } else {
            return "Withdrawal amount must be positive.";
        }
    }
}


let account1 = new BankAccount(500);
let account2 = new BankAccount(1000);

console.log(account1.deposit(200)); 

console.log(account1.withdraw(50)); 

console.log(account2.transfer(100, account1)); 
console.log("Account 1 Balance:", account1.getBalance()); // Expected: 750
console.log("Account 2 Balance:", account2.getBalance()); // Expected: 900
```
